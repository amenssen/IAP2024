```python
import numpy as np
from qutip import basis, mesolve, qeye, sigmax, sigmaz, Bloch
%matplotlib qt

# Input parameters
Omega = 1.0  # Rabi frequency (GHz, for simplicity in units)
delta = 0.0  # Detuning (GHz)
pulse_width = 5.0  # Pulse width (ns)

# Define time vector
t = np.linspace(0, pulse_width, 1000)  # Time from 0 to pulse width

# Hamiltonian of the system
H = -delta/2 * sigmaz() + Omega/2 * sigmax()  # Hamiltonian including detuning and driving term

# Initial state (ground state)
psi0 = basis(2, 0)

# Solve the time-dependent Schrodinger equation for the system
result = mesolve(H, psi0, t, [], [])

# Initialize the Bloch sphere
b = Bloch()

# Process each state at each time point for plotting on the Bloch sphere
for state in result.states:
    # The state is already a qutip.Qobj, so we can directly use it
    b.add_states(state)

# Show the Bloch sphere with the trajectory
b.show()
```
